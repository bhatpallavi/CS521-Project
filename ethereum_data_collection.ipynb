```python
#Import libraries for scraping
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup as soup
```

```python
#Function to extract top 100 influential crypto handles and return two lists
def getTwitterHandles():
    # Used cryptoweekly to get top 100 influencers
    url = "https://cryptoweekly.co/100/"

    # Retreives and parses page html
    req = Request(url, headers={'User-Agent': 'Mozilla/6.0'})
    client = urlopen(req, timeout=10)
    pageHtml = client.read()
    pageSoup = soup(pageHtml, "html.parser")
    print(pageSoup)
    profiles = pageSoup.findAll("div", {"class":"twitte crypto-text-color-blue"})
    print(profiles)
    twitterHandles = []
    twitter_emails = []
    print(profiles)
    for person in profiles:
        twitterHandles.append(person.a.string[1:])
        twitter_emails.append(person.a.get('href'))



    client.close()
    return twitterHandles, twitter_emails
```

```python
#Write crypto handles and emails into two text files
file_handles = open("twitter_names.txt","w")
file_emails = open("twitter_emails.txt","w")
twitter_handles , twitter_emails = getTwitterHandles()
for i in twitter_handles:
    file_handles.write(i + "\n")
for i in twitter_emails:
    file_emails.write(i + "\n")
```
